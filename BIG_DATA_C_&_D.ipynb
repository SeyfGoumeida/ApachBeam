## assign_cluster :

```python
centroïdes = {
  0.0: 5.626434077616644,
  1.0: 9.669327878423491
}
#Assigner à chaque point son numéro de cluster :
def assign_cluster(x, centroids):
  num_cluster_0 = list(centroids.keys())[0]
  num_cluster_1 = list(centroids.keys())[1]
  coord_center_0 = centroids.get(num_cluster_0)
  coord_center_1 = centroids.get(num_cluster_1)

  dist = lambda x, center: np.sum(np.sqrt((x - center)**2))
  dist0 = dist(x, coord_center_0)
  dist1 = dist(x, coord_center_1)

  if (dist0<dist1):
    num_cluster = num_cluster_0
  else:
    num_cluster = num_cluster_1

  return (num_cluster,x) 
assign_result=[]
#----------------TEST--------------------------
#Assigner à chaque point son numéro de cluster :
for i,j,k in df.values :
  assign_result.append(assign_cluster(i,centroïdes))
print(assign_result)
```

```python
class GetCentres(beam.DoFn):
  def process(self, data):
    centers_cluster0 = np.mean(data[np.where(data[:,0] == 0), 1])
    centers_cluster1 = np.mean(data[np.where(data[:,0] == 1), 1])
    dict = {(0, centers_cluster0), (1, centers_cluster1)}
    return dict
# ------------ Read data -------------------- 
f = open('DataSet.csv', 'r')
for lines in f:
  data = f.readlines()
# ------------ pipeline --------------------
with beam.Pipeline() as pipeline:
  
  centroids = (
      pipeline
      | "read" >> beam.Create(data)
      | beam.Map(lambda x: x.split(";"))
      | beam.Map(lambda x : float(x[0]))
      | beam.Map(lambda x:(randint(0,1), x))
      | beam.GroupByKey()
      | beam.Map(lambda x:(x[0], np.mean(x[1])))
      #| "print" >> beam.Map(print)
  )

  cluster = (
      pipeline 
      | "read_cluster" >> beam.Create(data)
      | beam.Map(lambda x: x.split(";"))
      | beam.Map(lambda x:float(x[0]))
      |'Assign Cluster' >> beam.Map(
          assign_cluster,
          centroids = beam.pvalue.AsDict(centroids)
          )
      | "second iteration" >> beam.GroupByKey()
      | "second centers" >> beam.Map(lambda x:(x[0], np.mean(x[1])))
      #| "print_cluster" >> beam.Map(print)
  )

  #----------------------KMEANS----------------------
  # le critére d'arrét à corriger
  i=1
  Kmeans=cluster
  while (i<20): #Kmeans_1==Kmeans

    Kmeans_1 = Kmeans
    i=i+1
    #print("--------------{}----------------".format(i))
    Kmeans = (
        pipeline 
        | "read{}".format(i) >> beam.Create(data)
        | "read{}a".format(i) >> beam.Map(lambda x: x.split(";"))
        | "read{}b".format(i) >> beam.Map(lambda x:float(x[0]))
        |'Assign Cluster {}'.format(i) >> beam.Map(
            assign_cluster,
            centroids = beam.pvalue.AsDict(Kmeans_1)
            )
        | "iteration{}".format(i) >> beam.GroupByKey()
        | "labels_centers{}".format(i) >> beam.Map(lambda x:(x[0], np.mean(x[1])))


    )
    if(Kmeans == Kmeans_1):
      break
  #--------------------Resultat -------------------------
  result = (
      pipeline 
      | "read{}".format(i+1) >> beam.Create(data)
      | "read{}a".format(i+1) >> beam.Map(lambda x: x.split(";"))
      | "read{}b".format(i+1) >> beam.Map(lambda x:float(x[0]))
      |'Assign Cluster {}'.format(i+1) >> beam.Map(
          assign_cluster,
          centroids = beam.pvalue.AsDict(Kmeans)
          )
      |"print result" >> beam.Map(print)
      #|  "write{}b".format(i) >> beam.io.WriteToText("result.csv")

  )

```

## Implimentation : un seul état

```python
import apache_beam as beam
from apache_beam.transforms.userstate import BagStateSpec

class MyClass(beam.DoFn):
  #Declaration des états
  Points_Cluster_1 = BagStateSpec(name='Points_Cluster_1',coder=beam.coders.PickleCoder())
  Centre_Cluster_1 = BagStateSpec(name='Centre_Cluster_1',coder=beam.coders.PickleCoder())
  Points_Cluster_2 = BagStateSpec(name='Points_Cluster_2',coder=beam.coders.PickleCoder())
  Centre_Cluster_2 = BagStateSpec(name='Centre_Cluster_2',coder=beam.coders.PickleCoder())
  def process(self, element,
              Points_Cluster_1=beam.DoFn.StateParam(Points_Cluster_1),
              Centre_Cluster_1=beam.DoFn.StateParam(Centre_Cluster_1),
              Points_Cluster_2=beam.DoFn.StateParam(Points_Cluster_2),
              Centre_Cluster_2=beam.DoFn.StateParam(Centre_Cluster_2),
              ):
    #Lecture des états
    list_Points_Cluster_1 =list(set(Points_Cluster_1.read()))
    list_Points_Cluster_2 =list(set(Points_Cluster_2.read()))
    list_Centre_Cluster_1 =list(set(Centre_Cluster_1.read()))
    list_Centre_Cluster_2 =list(set(Centre_Cluster_2.read()))
    #Initialisation des deux centre avec les deux premiers points
    key, value = element
    if len(list_Points_Cluster_1)==0 :
      Points_Cluster_1.add(float(value))
      Centre_Cluster_1.add(float(value))
    elif len(list_Points_Cluster_2)==0:
      Points_Cluster_2.add(float(value))
      Centre_Cluster_2.add(float(value))
    else:
      #Calculer le plus proche centre
      dist = lambda x, center: np.sum(np.sqrt((x - center)**2))
      dist0 = dist(value, list_Centre_Cluster_1[0])
      dist1 = dist(value, list_Centre_Cluster_2[0])
      #Comparaison & esaignement
      if (dist0<dist1):
        #Affectation du point au centre le plus proche
        Points_Cluster_1.add(float(value))
        #Recupiration de l'ensemble des points 
        list_numbers =list(set(Points_Cluster_1.read()))
        #Recalcule du centre
        calculate_centers = lambda list_numbers:(np.mean(np.array(list_numbers)))
        center = calculate_centers(list_numbers)
        #Mise à jour des centres
        Centre_Cluster_1.clear()
        Centre_Cluster_1.add(float(center))
      else:
        #Affectation du point au centre le plus proche
        Points_Cluster_2.add(float(value))
        #Recupiration de l'ensemble des points 
        list_numbers =list(set(Points_Cluster_2.read()))
        #Recalcule du centre
        calculate_centers = lambda list_numbers:(np.mean(np.array(list_numbers)))
        center = calculate_centers(list_numbers)
        #Mise à jour des centres
        Centre_Cluster_2.clear()
        Centre_Cluster_2.add(float(center))      

    list_Points_Cluster_1 =list(set(Points_Cluster_1.read()))
    list_Points_Cluster_2 =list(set(Points_Cluster_2.read()))
    list_Centre_Cluster_1 =list(set(Centre_Cluster_1.read()))
    list_Centre_Cluster_2 =list(set(Centre_Cluster_2.read()))

    #Affichage du process 
    print("Cluster: 1")
    print("Centre: ",list_Centre_Cluster_1,"| size: " ,len(list_Points_Cluster_1),"| Points :", list_Points_Cluster_1)
    print("---------------------------------")
    print("Cluster: 2")
    print("Centre: ",list_Centre_Cluster_2,"| size: " ,len(list_Points_Cluster_2),"| Points :", list_Points_Cluster_2)
    print("")
    print("**********************************")

    yield (list_Centre_Cluster_1,list_Points_Cluster_1),(list_Centre_Cluster_2,list_Points_Cluster_2)

# ------------ Read data -------------------- 
f = open('DataSet.csv', 'r')
for lines in f:
  data = f.readlines()
# ------------ pipeline --------------------
with beam.Pipeline() as pipeline:
  dataCREATION = (
      pipeline
      | "read" >> beam.Create(data)
      | beam.Map(lambda x: x.split(";"))
      | beam.Map(lambda x : float(x[0]))
      | beam.Map(lambda x:(1, x))
      #| "print" >> beam.Map(print)
  )
  (dataCREATION
    | "Update state" >> beam.ParDo(MyClass())
    #| "Print" >> beam.Map(print)
  )
```

```python
import apache_beam as beam
from apache_beam.transforms.userstate import BagStateSpec

class MyClass(beam.DoFn):
  state = BagStateSpec(name='numbers',coder=beam.coders.PickleCoder())
  state2 = BagStateSpec(name='centres',coder=beam.coders.PickleCoder())

  def process(self, element, numbers=beam.DoFn.StateParam(state),centres=beam.DoFn.StateParam(state2)):
    key, value = element
    #Ajouter le point(elemnt)
    numbers.add(float(value))
    #Recupiration de l'ensemble des points 
    list_numbers =list(set(numbers.read()))
    #Recalcule du centre
    calculate_centers = lambda list_numbers:(np.mean(np.array(list_numbers)))
    center = calculate_centers(list_numbers)
    #Mise à jour des centres
    centres.clear()
    centres.add(float(center))
    #Recupiration du centre
    list_centres = list(set(centres.read()))

    print("Cluster :",key,"| Centre :",list_centres[0],"| size : " ,len(set(numbers.read())),"| Points :", set(numbers.read()))
    yield key,center, set(numbers.read())

# ------------ Read data -------------------- 
f = open('DataSet.csv', 'r')
for lines in f:
  data = f.readlines()
# ------------ pipeline --------------------
with beam.Pipeline() as pipeline:
  dataCREATION = (
      pipeline
      | "read" >> beam.Create(data)
      | beam.Map(lambda x: x.split(";"))
      | beam.Map(lambda x : float(x[0]))
      | beam.Map(lambda x:(randint(0,1), x))
      #| "print" >> beam.Map(print)
  )
  (dataCREATION
    | "Update state" >> beam.ParDo(MyClass())
    #| "Print" >> beam.Map(print)
  )
```
